5. (9б). Створіть клас 3D-фігура без атрибутів, з методом get_volume для отримання об'єму фігури, що повертає 0 та __add __, який приймає self та other в якості аргументів, а повертає суму об'ємів 3d-фігур self та other.

```python
class Shape3D:
  def get_volume(self):
    return 0;

  def __add__(self, other):
    return self.get_volume() + other.get_volume()
```

6. (11б). Створіть 2 дочірніх класи від 3d-фігура: Циліндр та Куб, які мають атрибути, необхідні для розрахунку об'єму. Визначте метод get_volume в кожному з них так, щоби вони розраховували об'єм у залежності від формули для кожного типу фігури. Створіть обʼєкт класу Циліндр та обʼєкт класу Куб. Виконайте операцію суми за допомогою оператора + між ними, аби знайти сумарний об'єм.

```python
import math

class Cylinder(Shape3D):
  def __init__(self, radius, height):
    self.radius = radius
    self.height = height

  def get_volume(self):
    volume = math.pi * self.radius**2 * self.height
    return volume

class Cube(Shape3D):
  def __init__(self, side):
    self.side = side

  def get_volume(self):
    volume = self.side ** 3
    return volume


cylinder = Cylinder(1, 2)
cube = Cube(3)
print(cylinder+cube)
```

8. (13б). Створіть клас EWallet з приватними атрибутами balance та id. Реалізуйте методи поповнення та зняття коштів, забезпечивши належну інкапсуляцію. Підказка: використовуйте декоратори property та setter.

```python
class EWallet:
  def __init__ (self, balance, id):
    self.__balance = max(0, balance)
    self.__id = id

  @property
  def balance(self):
    return self.__balance

  @balance.setter
  def balance(self, value):
      if value >= 0:
          self.__balance = value

  def top_up(self, amound):
    self.balance += amound

  def withdraw(self, amound):
    if self.balance >= amound:
      self.balance -= amound
    else:
      print("Not enough money")
```

9. (11б). Створіть клас ShoppingCart, який містить список об'єктів типу Item. Реалізуйте методи для додавання айтемів, їх видалення та відображення списку айтемів.

```python
class Item:
  def __init__(self, name, price):
    self.name = name
    self.price = price

class ShoppingCart:
  def __init__(self):
    self.items = []

  def add_item(self, item):
    if isinstance(item, Item):
        self.items.append(item)

  def delete_item(self, item):
    if item in self.items:
      self.items.remove(item)

  def show_items(self):
    if not self.items:
        print("List is empty")
    else:
        for i in self.items:
            print(f"{i.name}: {i.price}")
```

В моєму розумінні, ітерабельний об'єкт - це об'єкт по якому можна пройти, відділяючи певний його вміст одне від одного, як сторінки в книзі, попри те, що це єдиний об'єкт (книга), а ітератор - це функція, що проходиться почергово по цим об'єктам, як закладинка чи читач, безпосередньо.

13. (11б). Створіть клас, що має визначений Iterator Protocol та при кожному виклику next повертає назву дня тижня, поки вони не вичерпаються. Зверніть увагу, що порядок має значення.

```python
class WeekDays:
    def __init__(self):
        self.days = [
            "Monday",
            "Tuesday",
            "Wednesday",
            "Thursday",
            "Friday",
            "Saturday",
            "Sunday"
        ]
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= len(self.days):
            raise StopIteration
        day = self.days[self.index]
        self.index += 1
        return day
```
