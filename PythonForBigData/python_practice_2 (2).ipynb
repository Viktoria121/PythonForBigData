15. Порівняйте список та кортеж. Назвіть схожості та відмінності, випадки використання.

І список і кортеж зберігають набір елементів одного або різних типів, дозволяють звертатися до елементів за їхнів порядковим номером.
Проте список дозволяє вносити корективи в уже існуючий об'єкт тобто є mutable, а для кортежа потрібно створювати новий об'єкт, тобто є immutable. Окрім цього список повільніший за кортеж і займає більше місця.
Список зазвичай використовується якщо дані в ньому потрібно буде змінювати, а кортеж, навпаки, коли потрібно зберігати без змін.

16. Створіть кортеж з 11ти елементів чисел з рухомою комою та отримайте кожен парний за індексом елемент в зворотньому порядку. Наприклад, маючи (1.2, 2.3, 3.3, 4.3, 5.3, 6.3, 7.3, 8.3, 9.3, 0.3), отримати (0.3, 8.3, 6.3, 4.3, 2.3). Результат збережіть в нову змінну та виведіть на екран.

```python
a=(1.2, 2.3, 3.3, 4.3, 5.3, 6.3, 7.3, 8.3, 9.3, 0.3)
b=tuple(reversed(a))
b=b[::2]
print(b)

c=a[::-2]
print(c)
```

19. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
a={1, 2, 3, 4, 5}
b={5, 6, 7, 8, 9}

c=a.union(b)
d=a.difference(b)
e=a.intersection(b)
f=a.symmetric_difference(b)

print(c, "\n", d, "\n", e, "\n", f)
```

# VII. Словники (Dictionary).

20. Створіть пустий словник. До нього додайте чотири пари елементів такі, щоб їхні ключі були різних типів. Чи може список бути ключем? Чому?

```python
dictionary={}
dictionary[1]=0
dictionary[0]=1
dictionary[True]="False"
dictionary["Yes"]="No"
dictionary[(1, True, "Yes")]="No, list cannot be dictionary"
print(dictionary)
```

Ні, список не може бути ключем, оскільки він є mutable. А ключі маєть бути незмінними.
